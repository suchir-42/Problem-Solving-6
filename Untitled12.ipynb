```python
# 1
def parking(capacity,logs):
  cur=0
  for i in logs:
    if i=='IN':
      cur+=1
    elif i=='OUT':
      if cur>0:
        cur-=1
  if cur>=capacity:
    status='Capacity Exceeded'
  else:
    status='Available'
  print('Currently Parked Vehicles:',cur)
  print('Parking Status:',status)
capacity=50
logs=["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking(capacity,logs)
```

```python
# 2
def delivery(distance,traffic,weather):
  speed=24
  base=(distance/speed)*60
  delay=0
  if traffic=='Low':
    delay+=5
  elif traffic=='Medium':
    delay+=10
  elif traffic=='High':
    delay+=20
  if weather=='Rainy':
    delay+=15
  edt=base+delay
  print(f'Estimated Delivery Time:{int(edt)} minutes')
delivery(8,'High','Rainy')
```

```python
# 3
def booking(total,booked):
  count=len(booked)
  occupancy=(count/total)*100
  if occupancy==100:
    status='Housefull'
  elif occupancy>=75:
    status='Almost Full'
  else:
    status='Available'
  print('Occupancy:',occupancy)
  print('Show Status:',status)
total=200
booked=[1]*150
booking(total,booked)
```

```python
# 4
def ServerLoad(cpu_readings):
  total = 0
  for i in cpu_readings:
    total += i
  avg = total / len(cpu_readings)
  if avg < 50:
    status = 'Normal'
  elif avg <= 80:
    status = 'Warning'
  else:
    status = 'Critical'
  print('Average CPU Load:', int(avg),'%')
  print('Server Status:', status)
cpu = [45, 60, 70, 85, 90]
ServerLoad(cpu)
```

```python
# 5
def monitorResources(usage):
  overused = []
  for resource in usage:
    if usage[resource] > 8:
      overused.append(resource)
  if overused:
    print('Overused Resources:',''.join(overused))
    print('Energy Alert: Yes')
  else:
    print('Overused Resources: None')
    print('Energy Alert: No')
usage = {
    'Projector': 6,
    'AC': 9,
    'Lights': 4
}
monitorResources(usage)
```

```python
# 6
def manageReg(capacity, registrations):
  if registrations > capacity:
    confirmed = capacity
    waitlist = registrations - capacity
    status = 'Closed'
  else:
    confirmed = registrations
    waitlist = 0
    status = 'Open'
  print('Confirmed Registrations:', confirmed)
  print('Waitlisted Users:', waitlist)
  print('Registration Status:', status)
manageReg(100, 105)
```
